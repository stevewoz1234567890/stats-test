## Fifa 2022 Web Scraping

```python
import numpy as np
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup as Soup
```

```python
url = "https://sofifa.com/players?offset="
```

```python
column = ['ID','Name','Full_Name','picture','Flag','Age','Position','Overall','Potential',
          'Team_Image','Team','Value']
```

```python
fifa2022_dataframe = pd.DataFrame(columns = column)
```

```python
fifa2022_dataframe
```

```python
for offset in range(0, 301): # 300 -> https://sofifa.com/players?offset=18000
    url = url + str(offset*60)
    print(url)
    p_html = requests.get(url)
    p_soup = p_html.text
    data = Soup(p_soup,'html.parser')
    table = data.find('tbody')
    for i in table.findAll('tr'):    
        td = i.findAll('td')
        picture = td[0].find('img').get('data-src')
        ID = td[0].find('img').get('id')
        flag = td[1].find('img').get('data-src')
        Name = td[1].findAll('a')[0].text
        Full_Name = td[1].findAll('a')[0]['data-tooltip']
        Age = td[2].text
        Position = td[1].findAll('a')[1].text
        Overall = td[3].find('span').text
        Potential = td[4].find('span').text
        Team_image = td[5].find('img').get('data-src')
        Team = td[5].find('a').text
        Value = td[6].text.strip()
        player_data = pd.DataFrame([[ID,Name,Full_Name,picture,flag,Age,Position,Overall,Potential,
                                 Team_image,Team,Value]])
        #print(player_data)
        player_data.columns = column
        fifa2022_dataframe = fifa2022_dataframe.append(player_data, ignore_index = True)
    url = url.replace(str(offset*60), '')  
```

```python
fifa2022_dataframe.drop_duplicates(inplace = True)
```

```python
fifa2022_dataframe
```

```python
# Get detailed player information from player page
detailed_columns = ['Preferred Foot', 'Best Position', 'Jersey Number', 'Height', 'Weight','Crossing', 'Finishing', 'Dribbling', 'Long Passing', 'Ball Control', 'Acceleration', 'Sprint Speed', 'Agility', 'Reactions', 'Balance', 'Shot Power', 'Jumping', 'Stamina', 'Strength', 'Aggression','Positioning', 'Defensive Awareness','ID']
detailed_data = pd.DataFrame(index = range(0, fifa2022_dataframe.count()[0]), columns = detailed_columns)
detailed_data.ID = fifa2022_dataframe.ID.values
```

```python
detailed_data
```

```python
player_data_url = 'https://sofifa.com/player/'
for id in fifa2022_dataframe.ID:
    print(id)
    url = player_data_url + str(id)
    source_code = requests.get(url)
    plain_text = source_code.text
    soup = Soup(plain_text, 'html.parser')
    skill_map = {}
    columns = soup.findAll('div', {'class': 'column col-4'})
    for column in columns:
        skills = column.findAll('li')
        for skill in skills:
            if(skill.find('label') != None):
                label = skill.find('label').text
                value = skill.text.replace(label, '').strip()
                skill_map[label] = value

    meta_data = soup.find('div', {'class': 'meta'}).text.split(' ')
    length = len(meta_data)
    weight = meta_data[length - 1]
    height = meta_data[length - 2].split('\'')[0] + '\'' + meta_data[length - 2].split('\'')[1]
    skill_map["Height"] = height
    skill_map['Weight'] = weight
    features = soup.findAll('ul', {'class': 'pl'})

    try:
        preferred_foot_label = features[0].find('label').text
        preferred_foot_all_text = features[0].find('li', {'class': 'bp3-text-overflow-ellipsis'}).text
        preferred_foot_value = preferred_foot_all_text.replace(preferred_foot_label, '').strip()
        skill_map[preferred_foot_label] = preferred_foot_value
    except:
        skill_map['Preferred Foot'] = ""

    try:    
        jersey_number_label = features[2].findAll('label')[1].text
        jersey_number_all_text = features[2].findAll('li')[2].text
        jersey_number_value = jersey_number_all_text.replace(jersey_number_label, '').strip()
        skill_map[jersey_number_label] = jersey_number_value
    except:
        skill_map['Jersey Number'] = ""

    try:
        crossing_all_text = features[3].findAll('li')[0].text.strip()
        crossing_value = crossing_all_text.split(" ")[0]
        crossing_label = crossing_all_text.split(" ")[1]
        skill_map[crossing_label] = crossing_value
    except:
        skill_map['Crossing'] = ""

    try:        
        finishing_all_text = features[3].findAll('li')[1].text.strip()
        finishing_value = finishing_all_text.split(" ")[0]
        finishing_label = finishing_all_text.split(" ")[1]
        skill_map[finishing_label] = finishing_value
    except:
        skill_map['Finishing'] = ""

    try:        
        dribbling_all_text = features[4].findAll('li')[0].text.strip()
        dribbling_value = dribbling_all_text.split(" ")[0]
        dribbling_label = dribbling_all_text.split(" ")[1]
        skill_map[dribbling_label] = dribbling_value
    except:
        skill_map['Dribbling'] = ""

    try:        
        long_passing_all_text = features[4].findAll('li')[3].text.strip()
        long_passing_value = long_passing_all_text.split(" ")[0]
        long_passing_label = long_passing_all_text.split(" ")[1]
        skill_map[long_passing_label] = long_passing_value
    except:
        skill_map['Long Passing'] = ""

    try:     
        ball_control_all_text = features[4].findAll('li')[4].text.strip()
        ball_control_value = ball_control_all_text.split(" ")[0]
        ball_control_label = ball_control_all_text.split(" ")[1]
        skill_map[ball_control_label] = ball_control_value
    except:
        skill_map['Ball Control'] = ""

    try:    
        acceleration_all_text = features[5].findAll('li')[0].text.strip()
        acceleration_value = acceleration_all_text.split(" ")[0]
        acceleration_label = acceleration_all_text.split(" ")[1]
        skill_map[acceleration_label] = acceleration_value
    except:
        skill_map['Acceleration'] = ""

    try:    
        sprint_speed_all_text = features[5].findAll('li')[1].text.strip()
        sprint_speed_value = sprint_speed_all_text.split(" ")[0]
        sprint_speed_label = sprint_speed_all_text.split(" ")[1]
        skill_map[sprint_speed_label] = sprint_speed_value
    except:
        skill_map['Sprint Speed'] = ""

    try:    
        agility_all_text = features[5].findAll('li')[2].text.strip()
        agility_value = agility_all_text.split(" ")[0]
        agility_label = agility_all_text.split(" ")[1]
        skill_map[agility_label] = agility_value
    except:
        skill_map['Agility'] = ""

    try:    
        reactions_all_text = features[5].findAll('li')[3].text.strip()
        reactions_value = reactions_all_text.split(" ")[0]
        reactions_label = reactions_all_text.split(" ")[1]
        skill_map[reactions_label] = reactions_value
    except:
        skill_map['Reactions'] = ""

    try:    
        balance_all_text = features[5].findAll('li')[4].text.strip()
        balance_value = balance_all_text.split(" ")[0]
        balance_label = balance_all_text.split(" ")[1]
        skill_map[balance_label] = balance_value
    except:
        skill_map['Balance'] = ""

    try:    
        shot_power_text = features[6].findAll('li')[0].text.strip()
        shot_power_value = shot_power_text.split(" ")[0]
        shot_power_label = shot_power_text.split(" ")[1]
        skill_map[shot_power_label] = shot_power_value
    except:
        skill_map['Shot Power'] = ""

    try:    
        jumping_text = features[6].findAll('li')[1].text.strip()
        jumping_value = jumping_text.split(" ")[0]
        jumping_label = jumping_text.split(" ")[1]
        skill_map[jumping_label] = jumping_value
    except:
        skill_map['Jumping'] = ""

    try:    
        stamina_text = features[6].findAll('li')[2].text.strip()
        stamina_value = stamina_text.split(" ")[0]
        stamina_label = stamina_text.split(" ")[1]
        skill_map[stamina_label] = stamina_value
    except:
        skill_map['Stamina'] = ""

    try:    
        strength_text = features[6].findAll('li')[3].text.strip()
        strength_value = strength_text.split(" ")[0]
        strength_label = strength_text.split(" ")[1]
        skill_map[strength_label] = strength_value
    except:
        skill_map['Strength'] = ""

    try:    
        aggression_text = features[7].findAll('li')[0].text.strip()
        aggression_value = aggression_text.split(" ")[0]
        aggression_label = aggression_text.split(" ")[1]
        skill_map[aggression_label] = aggression_value
    except:
        skill_map['Aggression'] = ""

    try:    
        positioning_text = features[7].findAll('li')[2].text.strip()
        positioning_value = positioning_text.split(" ")[0]
        positioning_label = positioning_text.split(" ")[1]
        skill_map[positioning_label] = positioning_value
    except:
        skill_map['Positioning'] = ""

    try:    
        defensive_awareness_text = features[8].findAll('li')[0].text.strip()
        defensive_awareness_value = defensive_awareness_text.split(" ")[0]
        defensive_awareness_label = defensive_awareness_text.split(" ")[1]
        skill_map[defensive_awareness_label] = defensive_awareness_value
    except:
        skill_map['Defensive Awareness'] = ""

    print(skill_map)    

    for key, value in skill_map.items():
        detailed_data.loc[detailed_data.ID == id, key] = value
```

```python
full_data = pd.merge(fifa2022_dataframe, detailed_data, how = 'inner', on = 'ID')
full_data.to_csv('files/fifa_2022_data.csv', encoding='utf-8-sig')
```
